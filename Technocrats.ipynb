```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
dataset = pd.read_csv("analytics.csv")
```

```python
dataset.head(10)
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
grouped_dataset = dataset.groupby('Delivery_Region')['Refunded_amount'].sum().reset_index()
plt.bar(grouped_dataset['Delivery_Region'], grouped_dataset['Refunded_amount'])
plt.xlabel('Delivery Region')
plt.ylabel('Refunded Amount')
plt.title('Analysis of Possiblity of Delivery Scams')
plt.show()
```

```python
plt.hist(dataset['Order_total'], bins=20)
plt.xlabel('Order Total')
plt.ylabel('Frequency')
plt.title('Histogram of Total Amount of the Order')
plt.show()
```

```python
plt.scatter(dataset['Order_total'], dataset['Refunded_amount'])
plt.xlabel('Order Total')
plt.ylabel('Refunded Amount')
plt.title('Order Total vs Refunded Amount')
plt.show()
```
